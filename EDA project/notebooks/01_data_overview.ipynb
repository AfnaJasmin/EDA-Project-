 ## Indian Road Accident Dataset (2022–2025)



 The Indian Road Accident Dataset (2022–2025) contains detailed information about road accidents that occurred across various cities and states in India. The dataset includes approximately 20,000 records with multiple features describing different aspects of each accident such as location, time, environmental conditions, and severity.

The dataset consists of both numerical and categorical variables. Numerical columns such as latitude, longitude, temperature, vehicles_involved, casualties, and risk_score provide quantitative insights, while categorical columns like city, state, weather, visibility, traffic_density, cause, and accident_severity describe the nature and conditions of accidents. The dataset also includes date and time-related features such as date, time, hour, day_of_week, and is_weekend, which are useful for time-based analysis.

This dataset is well-suited for Exploratory Data Analysis (EDA) to understand accident patterns, identify high-risk locations, analyze the impact of weather and traffic conditions, and study trends over time. It also enables detection of missing values, data quality issues, and supports feature engineering for deeper insights into accident risk factors.
